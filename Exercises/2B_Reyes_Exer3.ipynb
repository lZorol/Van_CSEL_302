```python
#@title Step 1: Importing Libraries
import pandas as pd #DataManipulation
import numpy as np  #NumericalOperation
import matplotlib.pyplot as plt #DataVisualization
from sklearn.linear_model import LinearRegression #DataModeling
```

```python
#@title Step 2: Generating Random Data

numP = 50

X = np.random.randint(0,100, numP)
Y = 2 * X + np.random.normal(0,10, numP)
print(f"X value: {X}")
print("")
print(f"Y value: {Y}")
```

```python
#@title Step 3 : Creating Data Frame and Saving to CSV

data = pd.DataFrame({'X': X, 'Y': Y})


data.to_csv('Delta_csv', index=False)

print(data.head(10))
```

```python
#@title Visualizing the Data
plt.scatter(data['X'],data['Y'])
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent variable (Y)')
plt.title('Scatter Plot of Data')
plt.show()
```

```python
#@title Step 5:  Preparing Data for Modeling
X = data [['X']]
Y = data [['Y']]
```

```python
#@title Step 6: Fitting the Linear Regression Model

model = LinearRegression()

model.fit(X,Y)
```

```python
#@title Step 7: Getting Model Coefficients

slope = model.coef_[0]

intercept = model.intercept_

print('Slope', slope)
print('Intercept', intercept)
```

```python
#@title Step 8: Visualizing the Linear Regression Line

plt.scatter(X,Y)
plt.plot(X,model.predict(X), color='pink')
plt.scatter(data['X'],data['Y'])
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent variable (Y)')
plt.title('Scatter Plot of Data')
plt.show()
```

```python
#@title Step 9; Making Prediction

new_X = [[np.random.randint(100,200)]]

predicted_Y = model.predict(new_X)

print('Value of X: ', new_X)
print('Predicted Value of Y: ', predicted_Y)
```
