```python
#@title Step 7: Feature and Important Analysis (Hours of Study)

x = DataFrame[['Hours_of_Study']]
y = DataFrame[['Grades']]

model = LinearRegression()
model.fit(x,y)

slope = model.coef_[0]
intercept = model.intercept_
print("Slope", slope)
print("Intercept", intercept)

plt.figure(facecolor = 'pink')
plt.scatter(x,y)
plt.plot(x, model.predict(x), color = 'lightblue', linewidth = 5)
plt.xlabel('Independent Variable', color = 'pink')
plt.ylabel('Dependent Variable', color = 'pink')
plt.title('Linear Regression')
plt.show()

new_x = [[np.random.randint(100,200)]]
predicted_y = model.predict(new_x)
print("New X", new_x)
print("Predicted Y for new X", predicted_y)
```
